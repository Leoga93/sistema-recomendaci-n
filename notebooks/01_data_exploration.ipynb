Toma de decisiones basada en los datos. se busca evidencia significativamente positiva de que los datos no siguen una distribución normal.

### Probabilidad de ajuste a la distribución normal

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

columns_numerics = ["Sales", "Profit", "Quantity", "Discount"]

rows = 2
cols = 2
fig, ax = plt.subplots(rows, cols, figsize=(12, 8))
ax = ax.flatten()

# Gráficos de la probabilidad de ajuste a la distribución normal
for i, col in enumerate(columns_numerics):
    stats.probplot(data[col], dist='norm', plot=ax[i])
    ax[i].set_title(f"Gráfico Q-Q para evaluar normalidad de {col}")
    ax[i].set_xlabel("Cuantiles teóricos")
    ax[i].set_ylabel("Cuantiles muestrales")
    
plt.tight_layout()
plt.show()
```

### Prueba de asimétria

```python
from scipy.stats import skewtest

# Hipótesis nula(H0), la asimetría de la población es igual a la de una distribución normal(0).
# Hipótesis relativa(H1), la asimetría de la población es diferente a la de una distribución normal.
# two-sided: bilateral, greater sesgo positivo(Cola a la derecha), less sesgo negativo(Cola a la izquierda)

valores_p_skewtest = {}

for col in columns_numerics:
    estadistica, valor_p = skewtest(a=data[col], axis=0, nan_policy="omit", alternative="greater")
    valores_p_skewtest[col] = valor_p
    
alpha = 0.05

for column, valor in valores_p_skewtest.items():
    if valor < alpha:
        print(f"Para {column}, se rechaza la hipótesis nula, los datos no tienen una distribución normal")
    else:
        print(f"Para {column}, se acepta la hipótesis nula, los datos tienen una distribución normal")
```

### Prueba de distribución de normalidad

```python
from scipy.stats import kstest

# Contraste de hipótesis
# Hipótesis nula (H0) = Los datos siguen una distribución normal
# Hipotesis relativa(H1) = Los datos no siguen una distribución normal
valores_p_kstest = {}

for col in columns_numerics:
    valores = kstest(rvs=data[col], cdf="norm", nan_policy="omit", alternative="two-sided")
    valores_p_kstest[col] = valores[1]

alpha = 0.05

for col, value in valores_p_kstest.items():
    if value < alpha:
        print(f"Para {col}:")
        print(f"El valor p({value}) es menor que alpha({alpha}) por lo que, se rechaza la hipótesis nula. Los datos no siguen una distribución normal")
        
    else:
        print(f"Para {col}:")
        print(f"El valor p{value} es mayor que alpha({alpha}) por lo que, se acepta la hipótesis nula. Los datos siguen una distribución normal")
```

### Conclusiones
* La columna Sales no sigue una distribución normal, ya que, presenta un sesgo positivo con altos valores atípicos a la derecha de la distribución. Se sugiere realizar una transformarción de datos logaritmica(Los valores están por encima de 0) y una normalización mediante el método robustScaler debido a la fuerte presencia de valores extremos.

* La columna Profit no sigue una distribución normal, ya que, presenta colas largas en los extremos y una curtosis leptocúrtica. Se sugiere realizar una transformarción de datos con yeo-johnson(Debido a la presencia de valores negativos) y una normalización mediante el método robustScaler debido a la fuerte presencia de valores extremos.

* La columna Discount, tiene valores discretos que ya se encuentran en un rango definido de 0 a 1. Se sugiere dejarla como esta o normalizar los datos entre 0 y 1.

* La columna Quantity, tiene valores discretos que ya se encuentran en un rango definido. Se sugiere normalizar los datos en un rando entre 0 y 1. Para dejar todas las variables en una escala semejate.

* Se hace evidente la gran acumulación de valores atípicos en los extremos en las característica 'Sales', 'Profit' y Shipping Cost. Se recomienda detectar y tratar los valores atípicos usando técnicas como el rango intercuartílico(IQR) o el Z-Score para identificar, y ajustar o eliminar los valores extremos.

### Calculos de rango intercuartil (IQR)

```python
# Calculando los cuartiles y IQR
Q1 = df_numeric.quantile(0.25)
Q3 = df_numeric.quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

sin_atipicos = pd.Series([True] * len(data))

for col in df_numeric.columns:
    valores_if = ((data[col] >= lower_bound[col]) & (data[col] <= upper_bound[col]))
    sin_atipicos = sin_atipicos & valores_if

df_sin_atipicos = data[sin_atipicos]
```
