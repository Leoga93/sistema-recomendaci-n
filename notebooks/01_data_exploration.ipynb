```python
# Definiendo las columnas y las filas 
rows, cols = 2, 4

# Creando la cuadricula y la figura
fig, ax = plt.subplots(rows, cols, figsize=(15, 15))
ax = ax.flatten()

for i, col in enumerate(data.select_dtypes(['int64', 'float64']).columns):
    sns.histplot(data[col], ax=ax[i], kde=True)
    ax[i].set_title(f"Histrograma de {col}")
    # Calcular la asimetría y curtosis
    skewness = data[col].skew() # Grado de asimetría, valor 0 distribución normal
    kurtosis = data[col].kurtosis() # grado de aputamiento, +++ leptocúrtica, --- platicúrtica
    ax[i].annotate(f'Skewness: {skewness:.2f}\nKurtosis: {kurtosis:.2f}', xy=(0.55, 0.90), xycoords='axes fraction', fontsize=10, color='red')

for j in range(len(data.select_dtypes(['int64', 'float64']).columns), len(ax)):
    ax[j].set_visible(False) # Oculta los gráficos vacíos
    
plt.tight_layout()
plt.show()
```

* Sales: Una asimetría alta de 8,14, lo que índica que la distribución esta fuertemente sesgada hacia la derecha y la mayoria de valores están en el extremo izquierdo, con pocos valores extremadamente altos que arrastran la media hacia la derecha y una curtosis extremadamente alta de 176,73 que índica que los valores estan inusualmente concentrados cerca de la media y colas muy largas, lo que siguiere la presencia de valores atípicos. Lo que podria sugerir errores en los datos o una distribución muy sesgada. Se sugiere una transformación de los datos, un cambio de escala (logaritmico, raíz cuadrada o Box-Cox) debido a que la asimetría tiene un valor extremadamente alto.

* Quantity: Una asimetría de 1,36 lo que sugiere una asimetría moderada. Se recomienda transformar los datos para normalizarlos.

* Discount: Una asimetría moderada con un valor de 1.39, se recomienda transformar los datos para normalizarlos. La curtosis tiene tendencia platicúrtica con un valor de 0.72.La distribución cuenta con valores extremos a la derecha.

* Profit: Una asimetría alta a la derecha con un valor de 4.16. La curtosis extremadamente alta con un valor de 291.41 sugiere que los valores estan inusualmente concentrados alrededor de la media y una concentración alta de valores atipícos en los extremos.
